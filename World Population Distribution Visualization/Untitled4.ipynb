```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
np.random.seed(0)
ages=np.random.randint(0,100,size=1000)             
plt.figure(figsize=(10,6))              
sns.histplot(ages,bins=10,kde=True)              
plt.title('Age Distribution of population')              
plt.xlabel('Age')              
plt.ylabel('number of people')              
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample manually created World Bank-like data
data = {
    'Country': ['India', 'China', 'USA', 'Indonesia', 'Brazil'],
    'Population_2022': [1410000000, 1420000000, 332000000, 276000000, 215000000]
}
df = pd.DataFrame(data)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Population_2022'], color='skyblue')
plt.title('Total Population by Country (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
